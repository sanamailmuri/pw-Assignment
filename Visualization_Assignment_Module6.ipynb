# BOKEH ASSIGNMENT:

#### Ques 1:- Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import numpy as np
from bokeh.plotting import figure, show

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
fig.line(x, y, legend_label="sin(x)")

show(fig)
```

#### Ques 2:- Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.models import ColumnDataSource

x, y = np.random.rand(2, 100)
sizes = np.random.randint(5, 20, 100)
colors = np.random.choice(['red', 'green', 'blue', 'yellow'], 100)
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

fig = figure(title="Scatter Plot", x_axis_label='x', y_axis_label='y')
fig.circle('x', 'y', size='sizes', fill_color='colors', source=source)


show(fig)
```

#### Ques 3:- Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
```

```python
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))
fig = figure(x_range=fruits, title="Fruit Counts", y_axis_label='Count')
fig.vbar(x='fruits', top='counts', width=0.5, source=source)
show(fig)
```

#### Ques 4:- Create a Bokeh histogram to visualize the distribution of the given data.

```python
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
```

```python
source = ColumnDataSource(data=dict(top=hist, left=edges[:-1], right=edges[1:]))

p = figure(title="Histogram")

p.quad(top='top', bottom=0, left='left', right='right', source=source)

show(p)
```
